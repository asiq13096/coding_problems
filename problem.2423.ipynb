<p style="text-align:center; color:#00D6F0;">We have a lowercase only string. We have to find all occurances of every letter in that string. Suppose we have a=2 , b=2 , c=1. So the string will be "aabbc". Here we can remove one letter and group of occurances will be equalized. Another example, given string "aaabbccdd". Here a=3 , b=2 , c=2 , d=2. If we remove one occurance from a, then all occurance will equalized. In these case our program should print True otherwise it will print False. Now solving this problem we have to approach step by step.</p>
<div style="text-align: center;">
    <ul style="display: inline-block; text-align: left; color:#F500F2;">
        <li>We have to convert given string to a key-value pair or dictionary in python. It will help us to compare occurances.</li>
        <li>Next we have to compare values only. There is one spcecial case added in this problem which is substracting one value.</li>
        <li>During comparing values we have to apply that special case. If we haven't this special case we may use set() function to compare easily.</li>
        <li>After comparing, we will get final result.</li>
    </ul>
</div>
<p style="text-align:center; color:#00D6F0;">There are two type of approach to create dictionary from given string. Either use built in Count() function or using loop to manually create a function. We will create two different source code for these two type of approach. Counter() belong to collections library. So we have to import it when we will use it.</p>

```python
#Using Counter()
from collections import Counter #Import counter to easily count elements in a collection
def equal_frequency(word):
    freq = Counter(word) #This will create a dictionary where every letter will be added as keys and their occurence will be as values
    freq_values = list(freq.values()) #Creating a new list using values only
    count = Counter(freq_values) #This counter will keep record in more precise way
    if len(count) == 1: #Case A: where All letter have same frequency
        return True
    elif len(count) == 2: #case B: There are exct two differeny frequency
        f1 , f2 = count.keys() #Extract thw two distinct frequency values
        if (count[f1] == 1 and (f1 - 1 == f2 or f1 - 1 == 0)) \
            or (count[f2] == 1 and (f2 - 1 == f1 or f2 - 1 == 0)):
            return True #Removal of one letter will equalize frequency
        else:
            return False #Not possible to equalize frequency by removing one letter
    else:
        return False #Case C: More than two different frequency exist
#Example Usage
words = ["abcc" , "aazze" , "aabbccddeee" , "AABBCDDD"]
for w in words:
    print(f"{w} -- > {equal_frequency(w)}")
```

```python
#Without using Counter()
def eq(word):
    frequency = {} #Initiate an empty dictionary to keep record
    for cha in word:
        if cha in frequency:
            frequency[cha] += 1 #If exist, increase one
        else:
            frequency[cha] = 1 #If not exist, Initialize
    #Building a list of values
    fr = list(frequency.values())
    frequen = {} #Another empty list to seperate occurances
    for c in fr:
        if c in frequen:
            frequen[c] += 1 # If exist increase
        else:
            frequen[c] = 1 #Initialize if doesn't exist
    #Building main logic
    if len(frequen) == 1: #Case A: All characters have the same frequency
        f_val = list(frequen.keys()) [0] #Extract that single value
        return f_val == 1  or len(frequency) == 1 # Return true if 1. frequency is 1 --> all characters appeared one
        #2. only one unique character in the string --> trivially equal frequency

    elif len(frequen) == 2: #Case B: Exactly two different frequencies
        f1 , f2 = frequen.keys() #Extracting two different frequency values
        #Case A: f1 occurs once and reducing it by 1 equals f2 or 0
        if frequen[f1] == 1 and (f1 - 1 == f2 or f1 - 1 == 0):
            return True
        #case B: f2 occurs once and reducing it by 1 equals f1 or 0
        if frequen[f2] == 1 and (f2 - 1 == f1 or f2 - 1 == 0):
            return True
        return False
    #Case C: More than two frequency --> Impossible
    else:
        return False
#Test Cases
words = ["dzwguwwjah" , "aabbcc" , "abcc" , "a" , "aaaa" , "aabbccc"]
for w in words:
    print(f"{w} ---> {eq(w)}")
```

```python
#Compact version
def eq(word):
    freq = {}
    for ch in word:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1
    for ch in list(freq.keys()):
        freq[ch] -= 1
        if freq[ch] == 0:
            del freq[ch]
        if len(freq) == 0 or len(set(freq.values())) == 1:
            return True
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1
    return False
words = ["abcc" , "aabbcc" , "aabbccc" , "a" , "abc" , "dzwguwwjah"]
for w in words:
    print(f"{w:10} --> {eq(w)}")
```
